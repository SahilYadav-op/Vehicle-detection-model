```python
from ultralytics import YOLO
import cv2
import numpy as np
```

```python
model = YOLO(r"../runs/detect/train7/weights/best.pt")
```

```python
cap = cv2.VideoCapture(r"C:\Users\sahil\Desktop\projects\road traffic test 1.mp4")

if not cap.isOpened():
    print("Error: Video file not found or cannot be opened.")
else:
    print("Video loaded successfully")
```

```python
line_y = 350
counted_ids = set()
vehicle_count = 0

for r in model.track(
        source=r"C:\Users\sahil\Desktop\projects\road test 3.mp4",
        stream=True,
        show=True):

    boxes = r.boxes

    if boxes is None:
        continue

    for box in boxes:

        if box.id is None:
            continue

        track_id = int(box.id[0])
        x1,y1,x2,y2 = map(int, box.xyxy[0])

        center_y = int((y1+y2)/2)

        if center_y > line_y and track_id not in counted_ids:
            counted_ids.add(track_id)
            vehicle_count += 1

    print("Vehicles counted:", vehicle_count)
```

```python
if vehicle_count > 25:
    print("Traffic Density: HIGH")

elif vehicle_count > 10:
    print("Traffic Density: MEDIUM")

else:
    print("Traffic Density: LOW")
```
